##Load the data

Because the test data from Kaggle doesn't have resulting sentiment score, we use train_test_split to split the training set into a train and a test set

##Load word vectors

```python
embed_df = pd.read_table('embeddings.txt', header=None, sep=' ', names=['word'] + range(1,51))
embeddings = {row[0]: list(row[1:]) for row in embed_df.as_matrix()}
```

##Transform the data

Create class that will tokenize and stem the corpus

```python
class StemTokenizer(object):
    def __init__(self):
        self.stemmer = PorterStemmer()
        self.tokenizer = RegexpTokenizer(r'\w+')
        
    def __call__(self, document):
        return [ self.stemmer.stem(token) for token in self.tokenizer.tokenize(document) ]
```

Create class that will vectorize a corpus using our StemTokenizer and word vectors

```python
class SentAnaVectorizer(TfidfVectorizer):
    def __init__(self, binary=False, use_idf=True, stop_words=u'english', analyzer=u'word', 
                 ngram_range=(1, 1), lowercase=True, max_df=0.5, min_df=1, max_features=None,
                 vocabulary=None, tokenizer=None, word_vectors=None):
        super(SentAnaVectorizer, self).__init__(self,
                                                binary=binary, use_idf=use_idf, stop_words=stop_words, analyzer=analyzer,
                                                ngram_range=ngram_range, lowercase=lowercase, max_df=max_df, min_df=min_df, 
                                                max_features=max_features, vocabulary=vocabulary,
                                                tokenizer=tokenizer)
        self.word_vectors = word_vectors
    
    def _phrase2vector(self, phrase):
        '''
        Converts a phrase into a single word vector by taking the mean (element-wise) of 
        the word vectors of the individual words of the phrase
        '''
        phrase_vector = [ [0 for i in range(1, 51)] ]
        
        #terms_list = re.findall(r"[\w']+|[!\"#$%&\'()*+,-./:;<=>?@\[\\\]^_`{|}~]", str(ngram).lower())
        words_list = re.findall(r"[\w']+", str(phrase).lower())
        for word in words_list:
            phrase_vector += [ self.word_vectors.get(word, [0 for i in range(1,51)]) ]
            
        return [ sum(x) for x in zip(*phrase_vector) ]

    def fit(self, X):
        
        # gets the list of brown vectors for each phrase in the train set
        self._vectors = {}
        for xi in X:
            self._vectors[xi] = self._phrase2vector(xi)
        
        return super(SentAnaVectorizer, self).fit(X)
        
    def transform(self, X):
        X_sparse = super(SentAnaVectorizer, self).transform(X)
        X_dense = X_sparse.todense()
        
        # transform using word vectors
        vectors_list = []
        for xi in X:
            v = self._vectors.get(xi, None)
            if not v:
                v = self._phrase2vector(xi)
            vectors_list += [ v ]
        vectors_array = np.array(vectors_list)
        
        # concatenate both results
        return np.concatenate((X_dense, vectors_array), axis=1)
 
    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)
```

Now actually transform the data

```python
vectorizer = SentAnaVectorizer(lowercase=False, word_vectors=embeddings)#, tokenizer=StemTokenizer())
```

```python
X_train_v = vectorizer.fit_transform(X_train)
```

Transform the test data

```python
X_test_v = vectorizer.transform(X_test)
X_test_v.shape
```

##Logistic Regression Model

Train the model

```python
classifier = LogisticRegression()
classifier.fit(X_train_v, y_train)
```

Evaluate the model

```python
# predict the y for the test data
predictions = classifier.predict(X_test_v)
```

```python
for i in range(10):
    print y_test[i], predictions[i]
    
print 'accuracy:\t', classifier.score(X_test_v, y_test)
print 'recall:\t\t', recall_score(y_test, predictions)
print 'precision:\t', precision_score(y_test, predictions)
print 'f1-score:\t', f1_score(y_test, predictions)
```
