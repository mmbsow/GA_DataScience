Import libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

Load all the data set

```python
data = pd.read_csv("sample_train.csv")
```

Split train data set into X and y variables and remove unused variables

```python
X = data.loc[:, [c for c in data.columns if c in ('current_coupon','time_to_maturity','is_callable','reporting_delay','trade_size','trade_type','curve_based_price')]]
y = data['trade_price']
```

Replace categorical columns

```python
X['non_callable'] = pd.Series([ 1 if i else 0 for i in(X['is_callable'] == 0) ])

X['sell'] = pd.Series([ 1 if i else 0 for i in(X['trade_type'] == 2) ])
X['buy'] = pd.Series([ 1 if i else 0 for i in(X['trade_type'] == 3) ])
X['dealer'] = pd.Series([ 1 if i else 0 for i in(X['trade_type'] == 4) ])

X = X.loc[:, [c for c in X.columns if c not in ('is_callable','trade_type')]]
```

```python
X.columns
```

Split X and y into training and test data sets

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
```

```python
print len(X_train)
print len(X_test)
```

Get statistics on explanatory variables for train data set

```python
X_train_df = pd.DataFrame(X_train, columns=X.columns)
y_train_df = pd.DataFrame(y_train, columns=['trade_price'])
X_train_df.describe()
```

```python
X_test_df = pd.DataFrame(X_test, columns=X.columns)
y_test_df = pd.DataFrame(y_test, columns=['trade_price'])
X_test_df.describe()
```

Plot relationship between y and explanatory variables

```python
#print 'Covariance <Trade Price, Time to Maturity> = ', y_train_df.cov(X_train_df['time_to_maturity'])
plt.scatter(X_train_df['time_to_maturity'], y_train_df)
plt.show()
```

## Mutliple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print regressor.intercept_
print regressor.coef_
```

```python
predictions = regressor.predict(X_test)
```

```python
for i,p in enumerate(predictions):
    print p, y_test[i]
```

```python
score = regressor.score(X_test, y_test)
print score
```

## Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
cubic_featurizer = PolynomialFeatures(degree=3)
```

```python
X_train_cubic = cubic_featurizer.fit_transform(X_train)
X_test_cubic = cubic_featurizer.transform(X_test)
```

```python
cubic_regressor = LinearRegression()
cubic_regressor.fit(X_train_cubic, y_train)
```

```python
cubic_regressor.score(X_test_cubic, y_test)
```
